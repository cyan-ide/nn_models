```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4)) # Adjust figsize as needed

llama_log = pd.read_csv('./logs/log.txt', sep=' ',header=None)
gpt_log = pd.read_csv('../gpt2/logs/log.txt', sep=' ',header=None)

llama_log.columns= ["step","eval_type","value"]
gpt_log.columns= ["step","eval_type","value"]

#validation loss
llama_loss = llama_log[ llama_log["eval_type"]=="val"]
gpt_loss = gpt_log[ gpt_log["eval_type"]=="val"]

ax1.plot(llama_loss['step'], llama_loss['value'], 'r', label='Llama2 Training Loss')
ax1.plot(gpt_loss['step'], gpt_loss['value'], 'g', label='GPT-2 Training Loss')

ax1.set_xlabel('steps')
ax1.set_ylabel('loss')
ax1.set_title('Training Loss (my Llama2 vs my GPT-2)')
ax1.grid(True)
ax1.legend()

#hellaswag accuracy
llama_hellaswag_acc = llama_log[ llama_log["eval_type"]=="hella"]
gpt_hellaswag_acc = gpt_log[ gpt_log["eval_type"]=="hella"]

ax2.plot(llama_hellaswag_acc['step'], llama_hellaswag_acc['value'], 'r', label='Llama2 Hellaswag acc')
ax2.plot(gpt_hellaswag_acc['step'], gpt_hellaswag_acc['value'], 'g', label='GPT-2 Hellaswag acc')

ax2.set_xlabel('steps')
ax2.set_ylabel('accuracy')
ax2.set_title('Hellaswag acc (my Llama2 vs my GPT-2)')
ax2.grid(True)
ax2.legend()

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```
